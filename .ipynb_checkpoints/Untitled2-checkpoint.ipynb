```python
from matplotlib.patches import Rectangle
import matplotlib.image as mpimg

def filter_candidate(image_id, candidates_pos, model, threshold):
    filtered_candidates_pos = []
    scores = []

    index = 0
    img = mpimg.imread(join("datasets/JPEGImages/{}.jpg".format(image_id)))
    for pos in candidates_pos:
        (x,y,w,h) = pos
        candidate = img[y:y+h, x:x+w]
        new_candidate = cv2.resize(candidate, (100, 100))
        (H, hogImage) = feature.hog(new_candidate, orientations=9, pixels_per_cell=(10, 10),
                                    cells_per_block=(2, 2), transform_sqrt=True, block_norm="L2-Hys", visualize=True)
        pred = model.predict_proba(H.reshape(1, -1))[0]

        if pred[0] >= threshold:
            filtered_candidates_pos.append([y, y+h, x, x+w])
            scores.append(pred[0])

        index += 1
    print(len(candidates_pos)-len(filtered_candidates_pos))
    return filtered_candidates_pos, scores

def load(filename):
    with open("model/{0}.pkl".format(filename), 'rb') as f:
        model = pickle.load(f)
    return model

def test_image(image_id, threshold, xmlfile):
    face_cascade = cv2.CascadeClassifier(xmlfile)

    fig, ax = plt.subplots(figsize=(20,20))
    ax.axis('off')

    img = mpimg.imread(join("datasets/JPEGImages/{}.jpg".format(image_id)))
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # 40-40 1.1 21
    faces = face_cascade.detectMultiScale(gray, 1.05, 15)
    
    hog_b = load("hog_balanced")
    new_faces, _ = filter_candidate(image_id, faces, hog_b, threshold)
    
    waldo = load("hog_waldo")
    waldo_faces, waldo_fscores = filter_candidate(image_id, new_faces, waldo, 0)
    wenda = load("svm_wenda")
    wenda_faces, wenda_fscores = filter_candidate(image_id, new_faces, wenda, 0)
    wizard = load("hog_wizard")
    wizard_faces, wizard_fscores = filter_candidate(image_id, new_faces, wizard, 0)

    im = ax.imshow(img)
    annos_waldo = get_annotation(image_id, target='waldo')
    for anno in annos_waldo:
        x_min, y_min, x_max, y_max = anno
        bbox = ax.add_patch(Rectangle((x_min,y_min),x_max-x_min,y_max-y_min, linewidth=1,
                                      edgecolor='b', facecolor=(1,1,1,0.7)))
        ax.text(x_min, y_min,'waldo')
    annos_wenda = get_annotation(image_id, target='wenda')
    for anno in annos_wenda:
        x_min, y_min, x_max, y_max = anno
        bbox = ax.add_patch(Rectangle((x_min,y_min),x_max-x_min,y_max-y_min, linewidth=1,
                                      edgecolor='b', facecolor=(1,1,1,0.5)))
        ax.text(x_min, y_min,'wenda')
    annos_wizard = get_annotation(image_id, target='wizard')
    for anno in annos_wizard:
        x_min, y_min, x_max, y_max = anno
        bbox = ax.add_patch(Rectangle((x_min,y_min),x_max-x_min,y_max-y_min, linewidth=1,
                                      edgecolor='b', facecolor=(1,1,1,0.5)))
        ax.text(x_min, y_min,'wizard')
    
    for y_min, y_max, x_min, x_max, score in zip(waldo_faces, waldo_fscores):
        bbox = ax.add_patch(Rectangle((x_min,y_min),x_max-x_min,y_max-y_min, linewidth=3,
                            edgecolor='r', facecolor=(1,1,1,0.3)))
        ax.text(x_min, y_min,'waldo-{}'.formate(score))
    for y_min, y_max, x_min, x_max, score in zip(wenda_faces, wenda_fscores):
        bbox = ax.add_patch(Rectangle((x_min,y_min),x_max-x_min,y_max-y_min, linewidth=3,
                            edgecolor='r', facecolor=(1,1,1,0.3)))
        ax.text(x_min, y_min,'wenda-{}'.formate(score))
    for y_min, y_max, x_min, x_max, score in zip(wizard_faces, wizard_fscores):
        bbox = ax.add_patch(Rectangle((x_min,y_min),x_max-x_min,y_max-y_min, linewidth=3,
                            edgecolor='r', facecolor=(1,1,1,0.3)))
        ax.text(x_min, y_min,'wizard-{}'.formate(score))
        
```
